# EDA for Store Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load dataset

```python
df = pd.read_csv('../data/staging/cleaned_stores.csv')
```

## Missing Values Heatmap

```python
plt.figure(figsize=(8,4))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap - Stores Data")
plt.show()
```

## Distribution of Store Sizes

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Size"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of Store Sizes")
plt.xlabel("Store Size")
plt.ylabel("Count")
plt.show()
```

## Store Type Counts

### For Type A

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Type_A", data=df, hue = "Type_A", legend = False, palette="Set2")
plt.title("Number of Stores by Type")
plt.xlabel("Store Type")
plt.ylabel("Count")
plt.show()
```

### For Type B

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Type_B", data=df, hue = "Type_B", legend = False, palette="Set2")
plt.title("Number of Stores by Type")
plt.xlabel("Store Type")
plt.ylabel("Count")
plt.show()
```

### For Type C

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Type_C", data=df, hue = "Type_C", legend = False, palette="Set2")
plt.title("Number of Stores by Type")
plt.xlabel("Store Type")
plt.ylabel("Count")
plt.show()
```

## Store Size by Type (Boxplot)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Type_A", y="Size", data=df, hue = "Type_A" , legend = False ,palette="pastel")
plt.title("Store Size Distribution by Store Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Type_B", y="Size", data=df, hue = "Type_B" , legend = False ,palette="pastel")
plt.title("Store Size Distribution by Store Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Type_C", y="Size", data=df, hue = "Type_C" , legend = False ,palette="pastel")
plt.title("Store Size Distribution by Store Type")
plt.show()
```

## Average Store Size by Type

```python
avg_size = df.groupby("Type_A")["Size"].mean()
avg_size.plot(kind="bar", figsize=(6,4), color="orange")
plt.title("Average Store Size by Type")
plt.ylabel("Average Size")
plt.show()
```

```python
avg_size = df.groupby("Type_B")["Size"].mean()
avg_size.plot(kind="bar", figsize=(6,4), color="orange")
plt.title("Average Store Size by Type")
plt.ylabel("Average Size")
plt.show()
```

```python
avg_size = df.groupby("Type_C")["Size"].mean()
avg_size.plot(kind="bar", figsize=(6,4), color="orange")
plt.title("Average Store Size by Type")
plt.ylabel("Average Size")
plt.show()
```
